Stack plot

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
area=[1,2,3,4,5]
area1=[1,6,3,4,2]
area2=[1,5,3,2,5]
l=["area","area1","area2"]
plt.stackplot(x, area,area1,area2, labels=l)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
area=[1,2,3,4,5]
area1=[1,6,3,4,2]
area2=[1,5,3,2,5]
l=["area","area1","area2"]
plt.stackplot(x, area,area1,area2, labels=l, colors=["y","b","r"])
plt.legend(loc=2)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("python")
plt.show()
```
