#Texts in Matplotlib

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,8]
y=[1,5,3,2,5]

plt.plot(x,y)

plt.title("information about title", fontsize=15)
plt.xlabel("Days",fontsize=11)
plt.ylabel("Python")

plt.text(2,5,"Python")

plt.show()
```
