```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7,8]
num=[2,3,1,4,5,6,7,4]

plt.scatter(day,num, color="r")
plt.xlabel("Day")
plt.ylabel("num")
plt.title("Sctter Plot", fontsize=20)
plt.show()
```

```python
day=[1,2,3,4,5,6,7,8]
num=[2,3,1,4,5,6,7,4]
colors=["r","y","b","m","g","b","y","k"]
size=[100,200,400,500,600,700,800,900]
plt.scatter(day,num, c=colors, s=size, alpha=0.5, marker="*", edgecolor="g", linewidth=2)
plt.xlabel("Day")
plt.ylabel("num")
plt.title("Sctter Plot", fontsize=20)
plt.show()
```

```python
day=[1,2,3,4,5,6,7,8]
num=[2,3,1,4,5,6,7,4]
colors=["r","y","b","m","g","b","y","k"]
size=[100,200,400,500,600,700,800,900]
plt.scatter(day,num, c=colors, s=size, alpha=0.5,cmap="viridis")
plt.colorbar()
plt.xlabel("Day")
plt.ylabel("num")
plt.title("Sctter Plot", fontsize=20)
plt.show()
```

```python
day=[1,2,3,4,5,6,7,8]
num=[2,3,1,4,5,6,7,4]
num2=[4,5,3,5,6,7,8,4]
colors=["r","y","b","m","g","b","y","k"]
size=[100,200,400,500,600,700,800,900]
plt.scatter(day,num, c=colors, s=size, alpha=0.5,cmap="BrBg")
plt.scatter(day,num2, color="r", s=size, alpha=0.5,)
t=plt.colorbar()
t.set_label("Color Bar")
plt.xlabel("Day")
plt.ylabel("num")
plt.title("Sctter Plot", fontsize=20)
plt.show()
```
