```python
import numpy as np
import matplotlib.pyplot as plt

x = ["py", "c", "c#", "java", "R"]
y = [89, 65, 56, 44, 78]
z = [20, 30, 40, 50, 60]
width = 0.2  # Width of bars

p = np.arange(len(x))  # Positions for the first bar
p1 = [j + width for j in p]  # Shifted positions for the second bar

plt.xlabel("Language", fontsize=10)
plt.ylabel("Number", fontsize=20)
plt.title("Bar Plot", fontsize=25, color="r")

plt.xticks(p + width / 2, x, rotation=10)  # Adjust x-axis labels to be centered

plt.bar(p, y, width, color="y", edgecolor="r", label="Popularity")
plt.bar(p1, z, width, color="r", edgecolor="r", label="Popularity1")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = ["py", "c", "c#", "java", "R"]
y = [89, 65, 56, 44, 78]
z = [20, 30, 40, 50, 60]
width = 0.2  # Width of bars

p = np.arange(len(x))  # Positions for the first bar
p1 = [j + width for j in p]  # Shifted positions for the second bar

plt.xlabel("Language", fontsize=10)
plt.ylabel("Number", fontsize=20)
plt.title("Bar Plot", fontsize=25, color="r")

plt.xticks(p + width / 2, x, rotation=10)  # Adjust x-axis labels to be centered

plt.barh(p, y, width, color="y", edgecolor="r", label="Popularity")
plt.barh(p1, z, width, color="r", edgecolor="r", label="Popularity1")

plt.legend()
plt.show()
```
