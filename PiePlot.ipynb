```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=True)
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False)
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False, radius=0.5)
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False, radius=0.5, labeldistance=1.2)
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False, radius=0.5, labeldistance=1.2, startangle=90)
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False, radius=0.5, labeldistance=1.2, startangle=180
       , textprops={"fontsize":8})
plt.show()
```

```python
x=[1,2,3,4,5]
y=["c","c++","java", "python","r"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=False, radius=0.5, labeldistance=1.2
       , textprops={"fontsize":8}, counterclock=False)
plt.show()
```

```python
x=[1,2,3,4,5]
x1=[40,30,20,10,60]
y=["c","c++","java", "python","r"]

c=['r','g','b','m','y']

plt.pie(x, labels=y, radius=1.5, colors=c)
plt.pie(x1, radius=0.5)
plt.show()
```

```python
x=[1,2,3,4,5]
x1=[40,30,20,10,60]
y=["c","c++","java", "python","r"]

c=['r','g','b','m','y']

plt.pie(x, labels=y, radius=1.5, colors=c)
plt.pie(x1, radius=0.5, colors="w")
plt.show()
```
