```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.8)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, labels=["python"])
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, labels=["python"], patch_artist=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x, labels=["python"], patch_artist=True, showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=["python"], patch_artist=True, showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
x1=[10,20,30,40,30,60,70, 90]

y=[x,x1]

plt.boxplot(y,labels=["python","c++"] ,patch_artist=True, showmeans=True)
plt.show()
```
